LIST OPERATIONS

```
#creating a list
lst = [2,3,4,5]
print(lst)
```

```
#concantenation
lst1 = [1,2,3,4]
lst2 = [5,6,7,8]
print(lst1+lst2)
```

```
#Repition
a = [2]*3
print(a)
```

```
#Accessing List Elements
b = [10,20,30,40,50,60]
print(b[0])
print(b[-1])
```

```
#List slicing
nums = [1,2,3,4,5,6,7,8]
print(nums[0:3])
print(nums[::-1])
```

```
#Modification
fruits = ["apple", "banana", "cherry"]
fruits[2] = "Orange"
print(fruits)
```

```
#Adding Elements
#1.append()
num = [2,5,3,6,7,8]
num.append(9)
print(num)

#2.extend()
num2 = [10,11,12,13]
num.extend(num2)
print(num)

#3.insert()
num.insert(2,4)
print(num)
```

```
#Deletin List items
#1.remove()
fruits = ["apple", "banana", "cherry"]
fruits.remove("banana")
print(fruits)

#2.pop()
fruits = ["apple", "banana", "cherry"]
removed_fruit = fruits.pop(1)
print(fruits)

#3.del()
fruits = ["apple", "banana", "cherry"]
del fruits[0]
print(fruits)

#4.clear()
fruits = ["apple", "banana", "cherry"]
fruits.clear()
print(fruits)
```

```
#List Comprehension
squares = [x**2 for x in range(1,11)]
print(squares)
```
